```python
from jikanpy import Jikan
jikan = Jikan()

print(jikan.anime(457)['data'])
```

```python
import sqlite3

conn = sqlite3.connect('database.db')
cursor = conn.cursor()
cursor.execute('''

CREATE TABLE IF NOT EXISTS anime (
               mal_id INTEGER PRIMARY_KEY,
               title TEXT NOT NULL,
               type TEXT,
               score DECIMAL,
               popularity INT,
               synopsis TEXT,
               url TEXT

               )
''')
```

```python

import time

def save_anime(data, cursor, connection):
    keys = ['mal_id', 'title', 'type', 'score', 'popularity', 'synopsis', 'url']

    try:

        cursor.execute('''
        INSERT INTO ANIME (mal_id, title, type, score, popularity, synopsis, url)
        VALUES (?, ?, ?, ?, ?, ?, ?)
''', tuple(data[key] for key in keys) )
        connection.commit()

    except sqlite3.IntegrityError:
        print("Skipping duplicate anime: " + data.get('title'))
    except Exception as e:
        print("Error: " + str(e))
```

```python


page = 1
next_page = True

while next_page:
    try:
        response = jikan.top('anime', page)


        for index, anime in enumerate(response['data']):

            save_anime(anime, cursor, conn)
            time.sleep(1.2)



        if response['pagination']['has_next_page'] == False:
            break

        print(f"Fetched page {page} with {len(response['data'])} anime")
    except Exception as e:
        print("Error2: " + str(e))

    page += 1


```
