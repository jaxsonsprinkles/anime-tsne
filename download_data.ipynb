```python
import sqlite3

conn = sqlite3.connect('database.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS anime (
               mal_id INTEGER PRIMARY_KEY,
               title TEXT NOT NULL,
               type TEXT,
               score DECIMAL,
               popularity INT,
               synopsis TEXT,
               url TEXT,
               image TEXT
               )
''')
```

```python
def save_anime(data, cursor, connection):
    keys = ['mal_id', 'title', 'type', 'score', 'popularity', 'synopsis', 'url', 'image']
    try:

        anime_tuple = tuple(data[key] for key in keys if key != 'image') 

        cursor.execute('''
        INSERT INTO ANIME (mal_id, title, type, score, popularity, synopsis, url, image) 
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
''', )
        connection.commit()
        print("Saved anime: " + data.get('title'))
    except sqlite3.IntegrityError:
        print("Skipping duplicate anime: " + data.get('title'))
    except Exception as e:
        print("Error: " + str(e))
   
```

```python
import time

page = 1
next_page = True

while next_page:
    try: 
        response = jikan.top('anime', page)
        

        for index, anime in enumerate(response['data']):
            save_anime(anime, cursor, conn)
            
            

        if response['pagination']['has_next_page'] == False:
            break

        print(f"Fetched page {page} with {len(response['data'])} anime")
    except Exception as e:
        print("Error2: " + str(e))

    page += 1


```
